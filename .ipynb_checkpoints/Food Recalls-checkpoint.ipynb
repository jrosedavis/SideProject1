```python
#Dependencies
%matplotlib inline
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
#Create file path to csv and read into csv
file_path = "data/recalls_US.csv"
recalls = pd.read_csv(file_path, encoding='cp1252')
```

```python
recalls.dtypes
```

```python
recalls
```

```python
#Change dtype of date column
recalls['Date'] = pd.to_datetime(recalls['Date'])
recalls['Date']
```

```python
#Filter date column to return values from 2019 only
start_date = "2019-1-1"
end_date = "2019-12-31"

after_start_date = recalls["Date"] >= start_date
before_end_date = recalls["Date"] <= end_date
between_two_dates = after_start_date & before_end_date
recalls_2019 = recalls.loc[between_two_dates]
len(recalls_2019)
```
